## Motivation  

This notebook illustrates the need to correct for the fluorescence decay along the Z-axis in confocal Z-stacks, as we described in [Saiz *et al.*, (2016) *JoVE*](https://www.ncbi.nlm.nih.gov/pubmed/26967230), and demonstrates the effect of such correction in the dataset at hand. 
We previously have applied this transformation in [Saiz *et al.*, (2016) *Nature Communications*](http://dx.doi.org/10.1038/ncomms13463) and in [Morgani *et al.*, (2018) *Developmental Biology*](https://doi.org/10.1016/j.ydbio.2018.06.017)

### Set up  

Load R packages we will need and preset some asthetics for plotting

````r
library('reshape2')
library('plyr')
library('dplyr')
library('ggplot2')

source('../src/plotting-aes.R')
````

### Load data  

Load interim data file ```new-lms-tx.csv```. This is a reference data set of embryos collected and fixed at different stages of blastocyst development and stained for several markers. This file is the output of the script ```new-lms_tx.R``` and its data is shown in Supplementary Figure 1b, d and e of the paper.

````r
new.lms <- read.csv('../data/interim/new-lms-tx.csv')
````

### Visualize uncorrected data  

Below are the uncorrected fluorescence values for the Trophectoderm marker CDX2, on log scale, along the Z-axis. Each panel shows data for all cells in all embryos of one litter.

````r
my.data <- new.lms %>% filter(Channel == 'CH2', Marker == 'CDX2.m', Treatment == 'Littermate')

ut.plot <- ggplot(data = my.data, 
                 aes(x = Z, y = log(CH2.Avg)))
ut.plot <- ut.plot + geom_jitter(aes(color = TE_ICM), 
                                 size = 1, alpha = 0.25) + 
      geom_smooth(color = 'black', size = 1, method = 'lm')
ut.plot <- ut.plot + 
      looks + 
      scale_color_manual(values = idcols) + 
      theme(aspect.ratio = 1/2, strip.text = element_blank()) + 
      labs(y = 'log[CDX2]', x = 'Z (µm)', color = 'TE vs ICM') + 
      facet_wrap( ~ Experiment) 
print(ut.plot)
````

It is quite evident that CDX2 levels in the TE in the above example decay along the Z axis in a quasi-linear fashion. This is not the case for the ICM. However, ICM cells generally do not express CDX2, and therefore any fit would only reflect a combination of background fluorescence and noise in the data. 
For the purposes of this example, I will not try to determine the quality of the linear fit. The fit is clearly imperfect, however, the correction method incorporates a step to account for population structure. The regression line shown fits a linear model for the entire population shown

### Visualize corrected data

Below is the previous plot, but showing the corrected values, which result from applying the custom function ```eb.cor()``` to the raw data. The values for ICM cells are clearly overcorrected, but we do not analyze that subset of data, so we will ignore it for the sake of this example. 

````r
my.data <- new.lms %>% filter(Channel == 'CH2', Marker == 'CDX2.m', Treatment == 'Littermate')

ut.plot <- ggplot(data = my.data, 
                 aes(x = Z, y = CH2.ebLogCor))
ut.plot <- ut.plot + geom_jitter(aes(color = TE_ICM), 
                                 size = 1, alpha = 0.25) + 
      geom_smooth(color = 'black', size = 1, method = 'lm')
ut.plot <- ut.plot + 
      looks + 
      scale_color_manual(values = idcols) + 
      theme(aspect.ratio = 1/2, strip.text = element_blank()) + 
      labs(y = 'log[CDX2], corrected', x = 'Z (µm)', color = 'TE vs ICM') + 
      facet_wrap( ~ Experiment) 
print(ut.plot)
````

### What about other markers?

The same approach works well for other markers we routinely use. Below is a comparable example for GATA6, the main Primitive Endoderm marker, which also shows a certain degree of expression in the Trophectoderm.  

First the uncorrected data:

````r
my.data <- new.lms %>% filter(Channel == 'CH5', Marker == 'GATA6.gt', Treatment == 'Littermate')

ut.plot <- ggplot(data = my.data, 
                 aes(x = Z, y = log(CH5.Avg)))
ut.plot <- ut.plot + geom_jitter(aes(color = TE_ICM), 
                                 size = 1, alpha = 0.25) + 
      geom_smooth(color = 'black', size = 1, method = 'lm')
ut.plot <- ut.plot + 
      looks + 
      scale_color_manual(values = idcols) + 
      theme(aspect.ratio = 1/2, strip.text = element_blank()) + 
      labs(y = 'log[GATA6]', x = 'Z (µm)', color = 'TE vs ICM') + 
      facet_wrap( ~ Experiment) 
print(ut.plot)
````

As in the previous example, there is a reasonable linear fit, illustrated by the regression line. In all datasets, TE cells outnumber ICM cells by about 2:1 or 3:1 ([Saiz *et al* (2016)](https://www.nature.com/articles/ncomms13463)). Therefore, the function ```eb.cor()``` does not distingush between TE and ICM cells, as TE cells make up the majority of the population and help reduce the noise.  

Below is the result of correcting GATA6 levels as previously

````r
my.data <- new.lms %>% filter(Channel == 'CH5', Marker == 'GATA6.gt', Treatment == 'Littermate')

ut.plot <- ggplot(data = my.data, 
                 aes(x = Z, y = CH5.ebLogCor))
ut.plot <- ut.plot + geom_jitter(aes(color = TE_ICM), 
                                 size = 1, alpha = 0.25) + 
      geom_smooth(color = 'black', size = 1, method = 'lm')
ut.plot <- ut.plot + 
      looks + 
      scale_color_manual(values = idcols) + 
      theme(aspect.ratio = 1/2, strip.text = element_blank()) + 
      labs(y = 'log[GATA6], corrected', x = 'Z (µm)', color = 'TE vs ICM') + 
      facet_wrap( ~ Experiment) 
print(ut.plot)
````

In this case we can see a near perfect correction of GATA6 levels, irrespective of the cell type (TE or ICM). Once again, the correcton has been done globally. Individual panels show values for a single litter of embryos purely to reduce overplotting.
